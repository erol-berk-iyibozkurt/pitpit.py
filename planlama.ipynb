film önerileri <br>

- Film puanlaması yaparak (bir uygulama gibi yaparsak, kullanıcı vb. olmalı) <br>
- Grafik şeklinde en çok algoritma olarak önerilen 5 film yatay sütun grafiği şeklinde sonuç versin.

Film önerme için en çok izlenen filmler içinden rastgele 10 filmi 10 üzerinden puanlarken film türü, vb ile bunun üzerinden 3 film önerme

## Film Önerme Algoritması

Puan bazlı 

- 70P Başlık benzerliği
- 25P aynı film türü
- 15P Aynı yönetmen
- 10P * her aynı oyuncu
- 10P seçilen filmden yüksek rating

Örnek:

- **Seçilen Film:** Star Wars 1

Star Wars 2 + 70P
aynı Genre + 25P
aynı yönetmen 15P
aynı oyuncular kaç tanes ise x 10P
daha yüksek puanlı ise 10P alır

Sonra yüzde olması için toplam puana filmin puanı bölünür.
